## Part 2: Transform the Data
Do we know that this client spent the more money than client 66037? If not, how would we find out? Transform the data using the steps below to prepare it for analysis.

```python
# Create a column that calculates the subtotal for each line using the unit_price and the qty
df['line_subtotal'] = df['unit_price'] * df['qty']
df.head()
```

```python
# Create a column for shipping price.

df['total_weight'] = df['unit_weight'] * df['qty']
df['shipping_price'] = df.apply(lambda row: row['total_weight'] * 7 if row['total_weight'] > 50 else row['total_weight'] * 10, axis=1)

df[['unit_price', 'unit_weight', 'qty', 'total_weight', 'shipping_price']]
```

```python
# Create a column for the total price using the subtotal and the shipping price along with a sales tax of 9.25%
df['line_price'] = (df['line_subtotal'] + df['shipping_price']) * 1.0925

df[['line_subtotal', 'shipping_price', 'line_price']].head(3)
```

```python
# Create a column for the cost of each line using unit cost, qty, and
# shipping price (assume the shipping cost is exactly what is charged to the client).
df['line_cost'] = (df['unit_cost'] * df['qty']) + df['shipping_price']
df.head(3)
```

```python
# Create a column for the profit of each line using line cost and line price
df['line_profit'] = df['line_price'] - df['line_cost']
df.head(3)
```

## Part 4: Summarize and Analyze
Use the new columns with confirmed values to find the following information.

```python
# How much did each of the top 5 clients by quantity spend? Check your work from Part 1 for client ids.
top_5_clients = df.groupby('client_id')['qty'].sum().nlargest(5)
print(top_5_clients)
top_5_clients_spend = df[df['client_id'].isin(top_5_clients.index)].groupby('client_id')['line_price'].sum()
print(top_5_clients_spend)
```

```python
# Create a summary DataFrame showing the totals for the for the top 5 clients with the following information:
# total units purchased, total shipping price, total revenue, and total profit. 
# Group the data by client_id and calculate the totals
client_totals = df.groupby('client_id').agg({
    'qty': 'sum',
    'shipping_price': 'sum',
    'line_price': 'sum',
    'line_cost': 'sum',
    'line_profit': 'sum'
})

# Get the top 5 clients by total units purchased
top_5_clients = client_totals['qty'].nlargest(5).index

# Filter the client_totals DataFrame to only include the top 5 clients
top_5_clients_totals = client_totals.loc[top_5_clients]

# Reset the index to display client_id as a column
top_5_clients_totals = top_5_clients_totals.reset_index()

# Print the result
print(top_5_clients_totals)
```

```python
# Format the data and rename the columns to names suitable for presentation.
# Group the data by client_id and calculate the totals
client_totals = df.groupby('client_id').agg({
    'qty': 'sum',
    'shipping_price': 'sum',
    'line_price': 'sum',
    'line_cost': 'sum',
    'line_profit': 'sum'
})

# Get the top 5 clients by total units purchased
top_5_clients = client_totals['qty'].nlargest(5).index

# Filter the client_totals DataFrame to only include the top 5 clients
top_5_clients_totals = client_totals.loc[top_5_clients]

# Reset the index to display client_id as a column
top_5_clients_totals = top_5_clients_totals.reset_index()

# Rename the columns
top_5_clients_totals.columns = [
    'Client ID', 
    'Total Units Purchased', 
    'Total Shipping Cost', 
    'Total Revenue', 
    'Total Cost', 
    'Total Profit'
]

# Format the data
top_5_clients_totals[['Total Shipping Cost', 'Total Revenue', 'Total Cost', 'Total Profit']] = top_5_clients_totals[['Total Shipping Cost', 'Total Revenue', 'Total Cost', 'Total Profit']].round(2)

# Print the result
print(top_5_clients_totals)
# Define the money columns. 

money_columns = ['Total Shipping Cost', 'Total Revenue', 'Total Cost', 'Total Profit']
# Define a function that converts a dollar amount to millions.
def dollars_to_millions(x):
    return x / 1e6

# Apply the currency_format_millions function to only the money columns.
top_5_clients_totals[money_columns] = top_5_clients_totals[money_columns].apply(dollars_to_millions)

# Rename the columns to reflect the change in the money format.
top_5_clients_totals.columns = [
    'Client ID', 
    'Total Units Purchased', 
    'Total Shipping Cost (M)', 
    'Total Revenue (M)', 
    'Total Cost (M)', 
    'Total Profit (M)'
]

top_5_clients_totals
```

```python
# Sort the updated data by "Total Profit (millions)" form highest to lowest and assign the sort to a new DatFrame.
```

```python
sorted_data = top_5_clients_totals.sort_values(by='Total Profit (M)', ascending=False)
sorted_data
```
